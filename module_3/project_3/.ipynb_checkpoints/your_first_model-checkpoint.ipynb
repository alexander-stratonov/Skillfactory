# Загрузка Pandas и очистка данных

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.model_selection import train_test_split
from ast import literal_eval

df = pd.read_csv('C:/Users/user/Documents/GitHub/Skillfactory_Alexander_Stratonov/module_3/project_3/main_task.csv')
```

```python
def price_rate(price):
    if price == "$":
        return 1
    elif price == "$$ - $$$":
        return 2
    elif price == "$$$$":
        return 3
    elif price == "":
        return NaN
df['Price Range'] = df['Price Range'].apply(price_rate)
```

```python
df[['Ranking','Rating','Price Range','Number of Reviews']]=df[['Ranking','Rating','Price Range','Number of Reviews']].fillna(df.mean())
df["Cuisine Style"] = df.groupby("City")["Cuisine Style"].transform(lambda x: x.fillna("['Other']"))
df['cuisin_count'] = df['Cuisine Style'].str.split(', ').str.len()
df["Cuisine Style"] = df["Cuisine Style"].apply(literal_eval)
df
```

# Разбиваем датафрейм на части, необходимые для обучения и тестирования модели

```python
df=df.drop(df.columns[[1,2,7,8,9]], axis=1)
```

```python
# Х - данные с информацией о ресторанах, у - целевая переменная (рейтинги ресторанов)
X = df.drop(['Restaurant_id', 'Rating'], axis = 1)
y = df['Rating']
```

```python
# Наборы данных с меткой "train" будут использоваться для обучения модели, "test" - для тестирования.
# Для тестирования мы будем использовать 25% от исходного датасета.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

# Создаём, обучаем и тестируем модель

```python
# Создаём модель
regr = RandomForestRegressor(n_estimators=100)

# Обучаем модель на тестовом наборе данных
regr.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = regr.predict(X_test)
```

```python
# Сравниваем предсказанные значения (y_pred) с реальными (y_test), и смотрим насколько они в среднем отличаются
# Метрика называется Mean Absolute Error (MAE) и показывает среднее отклонение предсказанных значений от фактических.
print('MAE:', round(metrics.mean_absolute_error(y_test, y_pred),5))

# 1rez = 0.42196, поменял Price Range
# 2rez = 0.41712, поменял пропуски в Cuisine Style на Other
# 3rez = 0.41309, добавил cuisin_count - количество кухни в ресторане
```
