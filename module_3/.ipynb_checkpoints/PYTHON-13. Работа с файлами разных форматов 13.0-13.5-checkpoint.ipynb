# 13.3 Группировка данных

```python
cuisines = []  # создаём пустой список для хранения уникальных значений кухонь
for recipe in recipes:  # начинаем перебор всех рецептов
    if not(recipe['cuisine'] in cuisines):  # если тип кухни текущего блюда ещё не встречался
        cuisines.append(recipe['cuisine']) # добавляем его к списку cuisines
print(len(cuisines))

cuisines = set()  # создаём пустое множество для хранения уникальных значений кухонь
for recipe in recipes:  # начинаем перебор всех рецептов
    cuisines.add(recipe['cuisine']) # добавляем название типа кухни к множеству
print(len(cuisines))
```

```python
all_ingredients= set()
for x in recipes:
    for y in x['ingredients']:
        all_ingredients.add(y)
len(all_ingredients)
```

```python
italian_ingredients= set()
for x in recipes:
    if x['cuisine'] == 'italian':
        for y in x['ingredients']:
            italian_ingredients.add(y)
len(italian_ingredients)
```

```python
russian_ingredients= set()
for x in recipes:
    if x['cuisine'] == 'russian':
        for y in x['ingredients']:
            russian_ingredients.add(y)
russian_ingredients
```

```python
food = {}
for x in all_ingredients:
    food[x] = 0
for y in recipes:
    for x in y['ingredients']:
        food[x] += 1
```

# 13.4 Создание датафрейма на основе JSON-файла

```python
import pandas as pd
df = pd.DataFrame(recipes)

def find_item(cell):
    if item in cell:
        return 1
    return 0

for item in all_ingredients:
    df[item] = df['ingredients'].apply(find_item)

df['ingredients'] = df['ingredients'].apply(len) 

df.to_csv('C:/Users/user/Documents/GitHub/Skillfactory_Alexander_Stratonov/module_3/data/recipes.csv', index = False)
```

# 13.5 Создание JSON-файла

```python
df = pd.read_csv('C:/Users/user/Documents/GitHub/Skillfactory_Alexander_Stratonov/module_3/data/recipes.csv')
df
```

```python
ids=list(df.id)
ids
```

```python
ingredients=list(df.columns[3:])
ingredients
```

```python
new_recipes = []
def make_list(x):
    return [y for y in ingredients if x[y].iloc[0] == 1]
for current_id in ids:
    cuisine = df[df['id'] == current_id]['cuisine'].iloc[0]
    current_ingredients = make_list(df[df['id'] == current_id])
    current_recipe = {'cuisine': cuisine, 'id': int(current_id), 'ingredients': current_ingredients}
    new_recipes.append(current_recipe)
new_recipes
```

```python
with open("C:/Users/user/Documents/GitHub/Skillfactory_Alexander_Stratonov/module_3/data/new_recipes.json", "w") as write_file:
    json.dump(new_recipes, write_file)
```
