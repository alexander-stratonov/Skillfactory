```python
log = pd.read_csv('C:/Users/user/Documents/GitHub/Skillfactory_Alexander_Stratonov/module_3/data/log.csv', header = None)
log.columns = ['user_id', 'time', 'bet', 'win']  
log = log.dropna(subset = ['time'])  
log['time'] = log['time'].apply(lambda x: x[1:])  
log['time'] = pd.to_datetime(log['time']) 
log['hour'] = log.time.apply(lambda x: x.hour)

def time_of_day(x):
    if x in range(0,6):
        return 'ночь'
    elif x in range(6,12):
        return 'утро'
    elif x in range(12,18):
        return 'день'
    elif x in range(18,24):
        return 'вечер'

log['hour'] = log['hour'].apply(lambda x: time_of_day(x))
log['hour'].value_counts(ascending=True)
```

```python
log = pd.read_csv('C:/Users/user/Documents/GitHub/Skillfactory_Alexander_Stratonov/module_3/data/log.csv', header = None)
log.columns = ['user_id', 'time', 'bet', 'win']  
log[['bet','win']] = log[['bet','win']].fillna(0)  

def fillna_win(row):
    if row.bet > 0 and row.win == 0:
        return -row.bet
    else:
        return row.win

new_win = log.apply(lambda row: fillna_win(row), axis=1)  
log['win'] = new_win
sum(log['win']<0)
```
