```python
def user_parser(user_id):
    if "#error" not in str(user_id):
        return str(user_id).replace("Запись пользователя № - ", "")
    else:
        return ''
        
def time_parser (time):
    if "[" in str(time):
        time=time[1::]
    return time

log.user_id = log.user_id.apply(user_parser)
log.time = log.time.apply(time_parser)

log = log[log['user_id']!='']
```
