```python
melb_data['Date'] = pd.to_datetime(melb_data['Date'])
```

```python
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car'] # список столбцов, которые мы не берём во внимание
for col in melb_data.columns: # цикл по именам столбцов
    if melb_data[col].nunique() < 150 and col not in cols_to_exclude: # проверяем условие
        melb_data[col] = melb_data[col].astype('category') # преобразуем тип столбца
display(melb_data.info())
```
