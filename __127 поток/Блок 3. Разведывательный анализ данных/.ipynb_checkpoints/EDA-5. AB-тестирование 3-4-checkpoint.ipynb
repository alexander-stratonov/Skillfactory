# 4. Анализ результатов A/B-тестирования: кумулятивные метрики

#### ЗНАКОМИМСЯ С ДАННЫМИ

```python
ab_data = pd.read_csv('data/ab_data.csv')
ab_data['timestamp'] = pd.to_datetime(ab_data['timestamp'], format='%Y-%m-%d')
ab_data
```

```python
daily_data = ab_data.groupby(['timestamp','group']).agg({
    'user_id':'count',
    'converted':'sum'
}).reset_index().rename(columns={'user_id': 'users_count'})
daily_data.head()
```

```python
daily_data['conversion'] = daily_data['converted'] / daily_data['users_count'] * 100
daily_data[(daily_data['timestamp'] == '2017-01-05') & (daily_data['group'] == 'A')]
```

#### ВИЗУАЛЬНЫЙ АНАЛИЗ КОНВЕРСИИ В ГРУППАХ

```python
# создаём фигуру размером 8x2
fig = plt.figure(figsize=(8, 2)) 
# добавляем систему координат
ax = fig.add_axes([1, 1, 1, 1]) 
# строим boxplot для conversion по признаку group
sns.boxplot(data=daily_data, x='conversion', y='group', ax=ax) 
# задаём подпись к графику
ax.set_title('Сравнение конверсии в группах')
# задаём отображение сетки
ax.grid();
```

```python
conversion_piv = daily_data.groupby('group')['conversion'].agg(['mean', 'median'])
conversion_piv
```

#### КУМУЛЯТИВНЫЕ МЕТРИКИ

```python
# создаём фигуру размером 8x2
fig = plt.figure(figsize=(8, 2))
# добавляем систему координат
ax = fig.add_axes([1, 1, 1, 1])
# строим lineplot для конверсии во времени в каждой группе
sns.lineplot(
    data=daily_data,
    x='timestamp',
    y='conversion',
    hue='group', 
    ax=ax
)
# задаём подпись к графику
ax.set_title('График конверсии по дням')
# задаём поворот меток на оси абсцисс
ax.xaxis.set_tick_params(rotation=30)
# задаём отображение сетки
ax.grid();
```

```python
# выделяем данные группы А
daily_data_a = daily_data[daily_data['group'] == 'A']
# считаем кумулятивное количество посетителей
daily_data_a.loc[:, 'cum_users_count'] = daily_data_a['users_count'].cumsum()
# выводим время, количество посетителей и кумулятивное количество посетителей
display(daily_data_a[['timestamp', 'users_count', 'cum_users_count']].head())
```

```python
# вычисляем кумулятивную сумму количества посетителей
daily_data['cum_users_count'] = daily_data.groupby(['group'])['users_count'].cumsum()
# вычисляем кумулятивную сумму количества совершённых целевых действий
daily_data['cum_converted'] = daily_data.groupby(['group'])['converted'].cumsum()
# вычисляем кумулятивную конверсию
daily_data['cum_conversion'] = daily_data['cum_converted']/daily_data['cum_users_count'] * 100
display(daily_data.head())
```

```python
# создаём фигуру размером 8x2
fig = plt.figure(figsize=(8, 2))
# добавляем систему координат
ax = fig.add_axes([1, 1, 1, 1])
# строим lineplot для кумулятивной конверсии во времени в каждой группе
sns.lineplot(x='timestamp', y='cum_conversion', data=daily_data, hue='group', ax=ax)
# задаём подпись к графику
ax.set_title('График кумулятивной конверсии по дням')
# задаём поворот меток на оси абсцисс
ax.xaxis.set_tick_params(rotation = 30)
# задаём отображение сетки
ax.grid(True);
```

### ЗАДАЧИ 4.6–4.9

```python
sample_a = pd.read_csv('data/ab_test-redesign_sample_a.csv')
sample_b = pd.read_csv('data/ab_test-redesign_sample_b.csv')
sample_a['date'] = pd.to_datetime(sample_a['date'], format='%Y-%m-%d')
sample_b['date'] = pd.to_datetime(sample_b['date'], format='%Y-%m-%d')
```

```python
sample_a.head(1)
```

```python
sample_b.head(1)
```

### 4.6

```python
sample_a1 = sample_a.groupby(['date']).agg({
    'cid':'count',
    'transactions':'sum',
    'revenue':'sum'
}).reset_index()

sample_b1 = sample_b.groupby(['date']).agg({
    'cid':'count',
    'transactions':'sum',
    'revenue':'sum'
}).reset_index()

sample_a1[sample_a1['date'] == '2020-11-08']
```

### 4.7

```python
sample_a1['cum_cid'] = sample_a1['cid'].cumsum()
sample_a1['cum_transactions'] = sample_a1['transactions'].cumsum()
sample_a1['cum_revenue'] = sample_a1['revenue'].cumsum()

sample_b1['cum_cid'] = sample_b1['cid'].cumsum()
sample_b1['cum_transactions'] = sample_b1['transactions'].cumsum()
sample_b1['cum_revenue'] = sample_b1['revenue'].cumsum()

sample_a1['convers'] = sample_a1['cum_transactions'] / sample_a1['cum_cid'] * 100
sample_b1['convers'] = sample_b1['cum_transactions'] / sample_a1['cum_cid'] * 100

sample_a1['re_tr'] = sample_a1['cum_revenue'] / sample_a1['cum_transactions']
sample_b1['re_tr'] = sample_b1['cum_revenue'] / sample_b1['cum_transactions']
```

```python
round(sample_a1['convers'].mean(), 2)
```

```python
round(sample_b1['convers'].mean(), 2)
```

### 4.8

```python
sample_a1['group'] = 'A'
sample_b1['group'] = 'B'
sample_a1b1 = pd.concat([sample_a1, sample_b1]).reset_index()
sample_a1b1 = sample_a1b1.drop(columns = ['index'], axis = 1)
```

```python
fig = plt.figure(figsize=(8, 2))
ax = fig.add_axes([1, 1, 1, 1])
sns.lineplot(x='date', y='convers', data=sample_a1b1, hue='group', ax=ax)
ax.set_title('График кумулятивной конверсии по дням')
ax.xaxis.set_tick_params(rotation = 30)
ax.grid(True);
```

### 4.9

```python
fig = plt.figure(figsize=(8, 2))
ax = fig.add_axes([1, 1, 1, 1])
sns.lineplot(x='date', y='re_tr', data=sample_a1b1, hue='group', ax=ax)
ax.set_title('График кумулятивного среднего чека')
ax.xaxis.set_tick_params(rotation = 30)
ax.grid(True);
```
