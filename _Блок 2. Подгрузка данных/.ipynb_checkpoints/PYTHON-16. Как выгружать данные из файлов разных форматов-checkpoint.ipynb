```python
from chardet.universaldetector import UniversalDetector # Импортируем субмодуль chardet.universaldetector
detector = UniversalDetector()
with open('C:/Users/user/Documents/GitHub/Skillfactory_Alexander_Stratonov/_Блок 2. Подгрузка данных/data/ErrorEnCoding.csv', 'rb') as fh:
    for line in fh:
        detector.feed(line)
        if detector.done:
            break
detector.close()
```
