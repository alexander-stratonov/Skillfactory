```python
import numpy as np
from IPython.display import display
import  pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv('C:/Users/user/Documents/GitHub/Skillfactory_Alexander_Stratonov/module_2/tips.csv')
display(df.total_bill.max())
```

```python
%matplotlib inline
df.plot()
```

```python
df['total_bill'].plot(kind = 'hist', grid = True, title = '  ')
```

```python
df['day'].value_counts().plot(kind = 'bar', 
                              grid = True, 
                              colormap = 'coolwarm', 
                              title = '   ')
```

```python
df.groupby(by = ['sex', 'smoker'])['tip'].mean().plot(kind = 'bar', grid= True)
```

```python
df[['total_bill', 'tip']].plot(kind = 'hist', 
                               grid = True, 
                               subplots = True, 
                               title = ['  ', ' '], 
                               legend = False)
```

```python
df.plot(x = 'total_bill', 
        y = 'tip', 
        kind = 'scatter', 
        grid = True, 
        title = '   Vs  ')
```

```python
df.pivot_table(values = ['total_bill', 'tip'], 
               index = 'day', 
               aggfunc = 'mean').plot(kind = 'bar')
```

```python
#  ,  ,      . 
df.pivot_table(values = 'tip', index = ['sex', 'smoker'], aggfunc = 'count').plot(kind = 'bar')
```

```python
#        ,    ,      ?
df.groupby('smoker')['tip'].mean().plot(kind = 'bar') 
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.hist(df['total_bill'])
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.hist(df['total_bill'], bins = 20, color = 'green')
axes.set_title('  ')
axes.set_ylabel(' ')
axes.set_xlabel(' , USD')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.scatter(x = df['total_bill'], y = df['tip'], marker = '8')
axes.set_title('   Vs C ')
axes.set_xlabel(' ')
axes.set_ylabel(' ')
```
